```python
import matplotlib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import pandas as pd
```

```python
path = 'E:\YandexDisk/DNAm draft/map_yakutia/'
coordinates_data_path = path
coordinates_data_file = 'coordinates.xlsx'
df = pd.read_excel(f"{coordinates_data_path}/{coordinates_data_file}")
longs = df['Longitude'].values
lats = df['Latitude'].values
colors = ['xkcd:' + df['Color'].values[i] for i in range(0, len(longs))]
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_global()
ax.stock_img()
ax.coastlines()
plt.scatter(x=longs,
            y=lats, 
            s=500, 
            c=colors,
            edgecolors='k',
            linewidth=1.5,
            transform=ccrs.PlateCarree())
ax.set_extent ((20, 150, 30, 83), ccrs.PlateCarree())
    
plt.savefig(f'{path}/map.png', bbox_inches='tight')
plt.savefig(f'{path}/map.pdf', bbox_inches='tight')
```
